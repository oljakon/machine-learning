```python
import numpy as np
import pandas as pd
import scipy
from sklearn.decomposition import PCA
```

```python
train = pd.read_csv('close_prices.csv')
train.head()
```

```python
pca = PCA(n_components=10)
```

```python
transformed_data = pca.fit_transform(train.iloc[:, 1:])
```

```python
features = pca.explained_variance_ratio_
print(features)
```

```python
i = 1
while np.sum(features[:i]) < 0.9:
    i += 1
print(i)
```

```python
file = open('421.txt', 'w')
file.write(str(i))
file.close()
```

```python
first_component = transformed_data[:, 0]
print(first_component)
```

```python
index = pd.read_csv('djia_index.csv')
corr_coef = np.corrcoef(first_component, index.iloc[:, 1])[0, 1]
print(corr_coef)
```

```python
file = open('422.txt', 'w')
file.write(str(round(corr_coef, 2)))
file.close()
```

```python
new_train = train.iloc[:, 1:]
company = new_train.keys()[pca.components_[0].argmax()]
print(company)
```

```python
file = open('423.txt', 'w')
file.write(company)
file.close()
```
